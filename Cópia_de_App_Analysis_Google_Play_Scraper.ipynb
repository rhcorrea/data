```
#Importando o comando app da library google-play-scraper
from google_play_scraper import app
```

```
#Verificando o app, utilizando seu id a partir da url da Google Play.

result = app(app_name, lang = 'sr')
result
```

```
#Baixando todos os reviews do app.
from google_play_scraper import Sort, reviews_all, reviews

Reviews = reviews_all( app_name, lang="sr", sort = Sort.NEWEST)
```

```
#Importando pandas.
import pandas as pd
```

```
#Transformando os dados em um DataFrame para trabalharmos as análises.
Reviews_LP = pd.DataFrame(Reviews).head(10000)
```
